```python
import gzip, traceback, glob,csv, os, time
```

```python
def splitMolBlock_exist(file):
    """
    .sdfまたは.sdf.gzファイルからCIDとSMILESを取得するジェネレータ。 https://zenn.dev/pipon_tech_blog/articles/616496074aefbb
    """
    gzip_file = True
    if file.endswith(".sdf.gz"):
        fi = gzip.open(file, 'r')
        gzip_file = True
    elif file.endswith(".sdf"):
        fi = open(file, 'r')
        gzip_file = False
    else:
        raise Exception("file extension is not sdf or sdf.gz")
    count = 0
    i = 1
    try:
        end_flag = True
        cid_flag = False
        smiles_flag = False
        for line in fi:
            # print(line)
            if gzip_file:
                line = line.decode('utf-8')

            if cid_flag:
                cid = int(line)
                cid_flag = False
            if smiles_flag:
                smiles = line.replace('\n', '')
                smiles_flag = False

            if 'PUBCHEM_COMPOUND_CID' in line:
                cid_flag = True

            if 'PUBCHEM_OPENEYE_CAN_SMILES' in line:
                smiles_flag = True

            if line[:4] == "$$$$":
                # if not cid in db_ids:
                yield cid, smiles
                count += 1
                i += 1

        if end_flag:
            #print(f"block {count} is done")
            yield cid, smiles
            #print("done")
    except Exception as e:
        print(traceback.format_exc())
        pass
    fi.close()
```

```python
file_path = glob.glob("./pubchem_downloads/CURRENT-Full_SDF/*sdf.gz")
print("sdf.gzファイル数",len(file_path))
print(file_path)
```

```python
csv_dir = "./pubchem_csv_full/"
os.makedirs(csv_dir, exist_ok=True)
j = 1
for file in file_path:
    with open(csv_dir + f'pubchem_all_compound_{j}.csv', 'w') as f:
        writer = csv.writer(f, lineterminator='\n') 
        writer.writerow(['CID', 'smiles']) 
        for row in splitMolBlock_exist(file):
            writer.writerow(row)
    j = j + 1
```

```python
csv_dir = "./pubchem_csv_full/"
os.makedirs(csv_dir, exist_ok=True)


with open(csv_dir + f'pubchem_all_compound_all.csv', 'w') as f:
    writer = csv.writer(f, lineterminator='\n') 
    writer.writerow(['CID', 'SMILES'])
    for file in file_path: 
        for row in splitMolBlock_exist(file):
            writer.writerow(row)
```

```python
import os
import csv

def count_total_rows_in_csvs(directory_path, include_header=True):
    total_rows = 0

    for filename in os.listdir(directory_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory_path, filename)
            with open(file_path, newline='', encoding='utf-8') as csvfile:
                reader = csv.reader(csvfile)
                rows = list(reader)
                row_count = len(rows)
                if not include_header and row_count > 0:
                    row_count -= 1
                print(f"{filename}: {row_count} rows")
                total_rows += row_count

    print(f"\nTotal rows across all CSV files: {total_rows}")
    return total_rows


total_rows = count_total_rows_in_csvs(csv_dir, include_header=False)
```
